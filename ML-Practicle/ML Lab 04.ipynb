```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier


raw_data = pd.read_csv('seeds.csv')


print(raw_data.info())

x=raw_data.iloc[:,:-1] #independent
y=raw_data.iloc[:,-1] #dependent


x_training_data, x_test_data, y_training_data, y_test_data = train_test_split(x, y,test_size=0.2,random_state=42)

print(x_test_data)



model = DecisionTreeClassifier(criterion = "entropy", random_state = 100, max_depth = 3, min_samples_leaf = 5, splitter='best')
model.fit(x_training_data,y_training_data)
predictions = model.predict(x_test_data)

print(predictions)
```
